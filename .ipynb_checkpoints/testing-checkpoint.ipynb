```python
data_business_activity = data[['niveaunaam', 'sdnaam', 'jaar', 'BHVEST', 'BHZZP', 'BHSTART', 'BHKANTRUIM', 'BHWP_1000INW']]
data_business_activity
# data_business_activity.isnull().sum()
```

```python
df_nulls = data_business_activity.loc[data_business_activity['BHVEST'].isnull()].index
data_business_activity_clean = data_business_activity.drop(df_nulls)
data_business_activity_clean
df_wrong_niveau = data_business_activity_clean.loc[data_business_activity_clean['niveaunaam'] != 'Stadsdelen'].index
data_business_activity_clean = data_business_activity_clean.drop(df_wrong_niveau)
data_business_activity_clean
```
